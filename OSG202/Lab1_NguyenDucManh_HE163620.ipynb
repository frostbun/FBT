#LAB – SHELL PROGRAMMING

###Lesson 1 : Print to the screen composite numbers <100. (Composite number is natural numbers greater than 1 and must be divisible by a number > 1 and itself.)

###Lesson 2 : Print out n first prime numbers (input n from keyboard).

###Lesson 3 : Print out the screen all prime numbers from 1000 to 2000.

###Lesson 4: Print out the screen the numbers <100 and divide by 5 or 6.

###Lesson 5: Print the first 4 perfect numbers on the screen (Perfect number is the number that equals the sum of its divisors including 1)

###Lesson 6 : In the natural numbers <= 100 count how many numbers there are:
 - Divide by 7.
 - Divide 7 with the remainder 2.
 - The end digit is “7”.

###Lesson 7: Given any natural number N> 1 (previously assigned). Print out the successful development of prime numbers from small to large.
Example:	 9 --> 3.3		12 --> 2.2.3

###Lesson 8: Given any natural number N (previously assigned). Print out all the different prime divisors of N.

###Lesson 9: Given an array of natural numbers, write a program that:
-	Arranges this array in descending order.
-	Print out the screen all prime numbers of this array.
-	Count how many prime numbers there are in the above array, how many composite numbers there are.

###Lesson 10: Given an array of integer numbers, delete the array with the elements = 0 and print out the rest of the array on the screen.

```python
%%shell

is_prime() {
    if(($1 < 2))
    then
        echo 0
        return
    fi
    for((i=2; i*i<=$1; ++i))
    do
        if(($1%i == 0))
        then
            echo 0
            return
        fi
    done
    echo 1
}

echo -n "Input n: "
read input
echo

# 1
echo -n "Composite numbers <100:"
for((i=1; i<100; ++i))
do
    if((!$(is_prime $i)))
    then
        echo -n " $i"
    fi
done
echo
echo

# 2
n=$input
echo -n "$n first prime numbers:"
for((i=2; n>0; ++i))
do
    if(($(is_prime $i)))
    then 
        echo -n " $i"
        ((--n))
    fi
done
echo
echo

# 3
echo -n "Prime numbers from 1000 to 2000:"
for((i=1000; i<=2000; ++i))
do
    if(($(is_prime $i)))
    then 
        echo -n " $i"
    fi
done
echo
echo

# 4
echo -n "The numbers <100 and divide by 5 or 6:"
for((i=0; i<100; ++i))
do
    if((i%5==0 || i%6==0))
    then
        echo -n " $i"
    fi
done
echo
echo

# 5
n=4
echo -n "$n frist perfect numbers:"
for((i=2; n>0; ++i))
do
    sum=1
    for((j=2; j*j<i; ++j))
    do
        if((i%j == 0))
        then
            ((sum += j + i/j))
        fi
    done
    if((j*j == i))
    then
        ((sum += j))
    fi
    if((sum == i))
    then
        echo -n " $i"
        ((--n))
    fi
done
echo
echo

# 6
n=100
echo "Natural numbers <= $n and divide by 7: $(((n+7) / 7))"
echo "Natural Numbers <= $n and divide 7 with remainder 2: $(((n+5) / 7))"
echo "Natural Numbers <= $n and the end digit is 7: $(((n+3) / 10))"
echo

# 7
n=$input
echo -n "$n --> "
for((i=2; i*i<=n; ++i))
do
    for((; n%i==0; n/=i))
    do
        echo -n "$i."
    done
done
if((n>1))
then
    echo "$n"
else
    printf "\b\n"
fi
echo

# 8
n=$input
echo -n "Different prime divisors of $n:"
for((i=2; i*i<=n; ++i))
do
    if((n%i == 0))
    then
        echo -n " $i"
        while((n%i==0))
        do
            ((n/=i))
        done
    fi
done
if((n>1))
then
    echo -n " $n"
fi
echo
echo

# 9
echo -n "Input array: "
read -a array
length=${#array[@]}
prime=""
count=0
for((i=0; i<length; ++i))
do
    if(($(is_prime ${array[i]})))
    then
        prime+=" ${array[i]}"
        ((++count))
    fi
    for((j=0; j<i; ++j))
    do
        if((array[i] > array[j]))
        then
            ((tmp = array[i]))
            ((array[i] = array[j]))
            ((array[j] = tmp))
        fi
    done
done
echo "Sorted in descending order array: ${array[@]}"
echo "Prime numbers in array:$prime"
echo "There are $count prime numbers and $((length-count)) composite numbers in array"
echo

# 10
echo -n "Input array: "
read -a array
length=${#array[@]}
for((i=0; i<length; ++i))
do
    if((!array[i]))
    then
        unset array[i]
    fi
done
array=(${array[@]})
echo "Output array: ${array[@]}"
```
